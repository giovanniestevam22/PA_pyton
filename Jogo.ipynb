```python
import random

palavras = ["Giovanni", "Henrique", "Evangelista", "Estevam",
            "Pedro", "Duarte", "Siqueira",
            "Arthur", "Carvalho", "Silva", "Vieira",
            "Eduardo", "Lemes", "Souza", "Luciano",
            "Lorenzo","Gabriel", "Greco", "Lopez",
            "Lucas", "Toshiharu", "Tsuyama",
            "Mulas", "IA" , "Musica", "Rock", "Sertanejo",
            "Biblia", "Comida", "Magro", "Cabelo", "Fusca",
            "psicodelismo", "lutas", "Futebol", "Cristo"]

chances = 25

while True:
    palavra = random.choice(palavras).lower()
    tentativa = input("Digite uma palavra  : ").lower()

    if tentativa == palavra:
        print("Parabéns! Você acertou a palavra!")
        break
    else:
        chances -= 1
        print(f"Você errou! Você tem {chances} chances restantes.")

    if chances <= 0:
        print("Você perdeu!!!")
        print("A palavra era: ", palavra)
        break
```
