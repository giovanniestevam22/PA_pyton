1- Numeros pares e impares em uma lista,
O Usuario digita uma sequencia de numeros separados por espaço, O Programa transforma a lista de inteiros., usando for, deve contar o numero de pares ou impares



```python
numeros = input().split()
pares = sum(1 for num in numeros if int(num) % 2 == 0)
impares = len(numeros) - pares
print(f"Pares: {pares}, Impares: {impares}")
```

2- Adivinhar um numero secreto, o programa escolhe um numero (ex:1 a 20), o usuario tenta adivinhar, enquanto não acertar, o programa da dicas(maior ou menor), quando acertar, o loop acaba, Dica: import random secreto =random.randint (1, 20) tentativa = none

```python
import random
secreto = random.randint(1, 10000)
chute = 0
tentativas = 15
while tentativas > 0 and chute != secreto:
    chute = int(input())
    if chute < secreto:
        print("Maior")
    elif chute > secreto:
        print("Menor")
    tentativas -= 1
if chute == secreto:
    print("Acertou!")
else:
    print("Tentativas esgotadas!")
```
