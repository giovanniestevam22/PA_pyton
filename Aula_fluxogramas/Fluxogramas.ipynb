```python
#Calcular Preço de Carro

##Entrada
fabrica = float(input("Digite o preço do carro fabricado: "))
perdistribuidor = float(input("Digite o percentual de distribuidor: "))
impostos = float(input("Digite o percentual de impostos: "))

##Processamento
preco = fabrica + (fabrica * (perdistribuidor/100)) + (fabrica * (impostos/100))

##Saída
print("O preço final do carro é: ", preco)
```

```python
#Conversão de horas e minutos em segundos

##Entrada
horas = int(input("Digite a quantidade de horas: "))
minutos = int(input("Digite a quantidade de minutos: "))

##Processamento
segundos = (horas * 3600) + (minutos * 60)

##Saída
print("O total de segundos é: ", segundos)
```

```python
#Calcular valor da hora extra

##Entrada

### Solicita as horas e minutos trabalhados separadamente
horas = int(input("Digite a quantidade de horas trabalhadas: "))
minutos = int(input("Digite a quantidade de minutos trabalhados: "))

### Solicita o valor da hora de trabalho
valor_hora = float(input("Digite o valor ganho por hora: "))

##Processamento

### Converte minutos para fração de hora
horas_totais = horas + (minutos / 60)

### Calcula o valor total a receber com acréscimo de 70%
valor_total = horas_totais * valor_hora * 1.7

##Saída
print(f"O total a receber pelas horas extras é: R$ {valor_total:.2f}")
```
