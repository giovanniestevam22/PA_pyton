```python
# Exercício 1 e 2 - Caixa Eletrônico com Login, Saque, Depósito e Limite
def caixa_eletronico():
    usuario_correto = "cliente"
    senha_correta = "1234"
    limite = 3000
    saldo = -50.00

    print("=== LOGIN ===")
    usuario = input("Usuário: ")
    senha = input("Senha: ")

    if usuario != usuario_correto or senha != senha_correta:
        print("Usuário ou senha incorretos. Tente novamente.")
        return

    while True:
        print("\n=== MENU ===")
        print("1 - Sacar")
        print("2 - Depositar")
        print("3 - Ver saldo")
        print("0 - Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            valor = float(input("Digite o valor para sacar: "))
            if saldo - valor >= -limite:
                saldo -= valor
                print(f"Saque autorizado. Novo saldo: R${saldo:.2f}")
                if saldo < 0:
                    print(f"Você está usando R${-saldo:.2f} do seu limite.")
            else:
                print("Saque negado. Limite insuficiente.")

        elif opcao == "2":
            valor = float(input("Digite o valor para depositar: "))
            saldo += valor
            print(f"Depósito realizado. Novo saldo: R${saldo:.2f}")
            if saldo > 0:
                print("Limite foi recuperado.")
            elif saldo > limite:
                print("Saldo excedeu o limite permitido.")

        elif opcao == "3":
            print(f"Saldo atual: R${saldo:.2f}")
            print(f"Limite restante: R${limite + saldo:.2f}")

        elif opcao == "0":
            print("Saindo...")
            break

        else:
            print("Opção inválida.")


# Exercício 3 - Competição de Vôlei
def pode_competir():
    altura = float(input("Digite sua altura (em metros): "))
    idade = int(input("Digite sua idade: "))
    if altura >= 1.75 and idade >= 18:
        print("Você pode competir.")
    else:
        print("Você NÃO pode competir.")


# Exercício 4 - Categoria de Boxe
def categoria_boxe():
    peso = float(input("Digite seu peso (kg): "))
    if peso <= 52.2:
        print("Categoria: Galo")
    elif peso <= 61.2:
        print("Categoria: Pena")
    elif peso <= 72.6:
        print("Categoria: Leve")
    elif peso <= 88.9:
        print("Categoria: Médio")
    else:
        print("Categoria: Pesado")


# Exercício 5 - Imposto de Renda
def imposto_renda():
    salario = float(input("Digite seu salário: "))
    if salario <= 1903.98:
        print("Isento de IR.")
    elif salario <= 2826.65:
        imposto = salario * 0.075
    elif salario <= 3751.05:
        imposto = salario * 0.15
    elif salario <= 4664.68:
        imposto = salario * 0.225
    else:
        imposto = salario * 0.275
    if salario > 1903.98:
        print(f"Você pagará R${imposto:.2f} de imposto de renda.")


# Exercício 6 - Aumento Salarial
def aumento_salarial():
    salario = float(input("Digite o salário atual: "))
    if salario <= 1500:
        aumento = salario * 0.20
    elif salario <= 2000:
        aumento = salario * 0.15
    elif salario <= 3000:
        aumento = salario * 0.10
    else:
        aumento = salario * 0.05
    novo_salario = salario + aumento
    print(f"Aumento: R${aumento:.2f}. Novo salário: R${novo_salario:.2f}")


# Exercício 7 - Simulador de Frete Inteligente
def simulador_frete():
    peso = float(input("Peso do pacote (kg): "))
    distancia = float(input("Distância (km): "))
    if peso <= 2:
        preco_por_100km = 5
    elif peso <= 5:
        preco_por_100km = 8
    else:
        preco_por_100km = 10

    frete = (distancia / 100) * preco_por_100km
    print(f"Para um pacote de {peso}kg viajando {distancia}km, o valor do frete será R${frete:.2f}")


# Menu Principal para executar cada exercício separadamente
def menu_principal():
    while True:
        print("\n=== ESCOLHA O EXERCÍCIO ===")
        print("1 - Caixa Eletrônico")
        print("2 - Competição de Vôlei")
        print("3 - Categoria de Boxe")
        print("4 - Imposto de Renda")
        print("5 - Aumento Salarial")
        print("6 - Simulador de Frete")
        print("0 - Sair")
        escolha = input("Opção: ")

        if escolha == "1":
            caixa_eletronico()
        elif escolha == "2":
            pode_competir()
        elif escolha == "3":
            categoria_boxe()
        elif escolha == "4":
            imposto_renda()
        elif escolha == "5":
            aumento_salarial()
        elif escolha == "6":
            simulador_frete()
        elif escolha == "0":
            print("Programa finalizado.")
            break
        else:
            print("Opção inválida!")

# Executar menu
menu_principal()
```
